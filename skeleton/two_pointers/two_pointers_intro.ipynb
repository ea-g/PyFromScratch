# The two-pointers technique  
Taking a look at using two "pointers" to solve problems more efficiently compared to double loops.   

## 1. The Problem with "Brute Force"

Imagine you have a sorted list of numbers and you want to find two numbers that add up to a specific target, $t$.  

**The Slow Way (Nested Loops):**  
You pick the first number, then check every other number. Then you pick the second number, and check every other number...
In computer science, we call this $O(n^2)$ complexity. If the list has 10,000 numbers, your computer might have to do 100,000,000 operations.

## 3. Why use this? (Speed Check)  

The Two-Pointer approach is $O(n)$. This means if you have 10,000 numbers, you only have to look at each number at most once.  

* **Brute Force:** 100,000,000 operations (worst case).  
* **Two-Pointers:** 10,000 operations (worst case).  

The Two-Pointer way is 10,000 times cheaper in operations!

```python
from random import randint, seed
import timeit

seed(55)

big_list = [randint(0, 9999) for i in range(10000)]
big_list.sort()
targ = randint(0, (9999*2)-1)
```

```python
s = timeit.default_timer()
print(find_sum_brute(big_list, targ))

print(f"Brute force took {timeit.default_timer() - s} seconds")
```

```python
s = timeit.default_timer()
print(find_sum_tp(big_list, targ))

print(f"Brute force took {timeit.default_timer() - s} seconds")
```
