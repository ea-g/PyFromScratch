# The two-pointers technique  
Taking a look at using two "pointers" to solve problems more efficiently compared to double loops.   

## 1. The Problem with "Brute Force"

Imagine you have a sorted list of numbers and you want to find two numbers that add up to a specific target, $t$.  

**The Slow Way (Nested Loops):**  
You pick the first number, then check every other number. Then you pick the second number, and check every other number...
In computer science, we call this $O(n^2)$ complexity. If the list has 10,000 numbers, your computer might have to do 100,000,000 operations.

## 2. The Two-Pointer Strategy  
 
If the list is **sorted**, we can be much smarter. We put one pointer at the *start (left)* and one at the *end (right)*.  

* **If the sum is too small:** Move the `left` pointer to the right (to get a bigger number).  
* **If the sum is too big:** Move the `right` pointer to the left (to get a smaller number).  
* **If the sum is perfect:** You found it! Return the two numbers. 

```python
def find_sum_brute(numbers: list[int], target:int) -> tuple[int]:
    # fill in the brute force method here (double loop)
    for i in range(len(numbers)):
        for j in range(i, len(numbers)):
            sm = numbers[i] + numbers[j]
            if sm == target:
                return (numbers[i], numbers[j])


def find_sum_tp(numbers: list[int], target:int) -> tuple[int]: 
    # fill in the two-pointer strategy here
    left = 0 
    right = len(numbers) - 1
    
    # we stop if the pointers reach each other
    while left < right:
        sm = numbers[left] + numbers[right]
        if sm == target:
            return (numbers[left], numbers[right])
        elif sm < target:
            left += 1
        else: 
            right -= 1
```

```python
# test case
nums = [1, 2, 4, 6, 8, 10, 11]
goal = 14
print(f"Result: {find_sum_brute(nums, goal)}")
print(f"Result: {find_sum_tp(nums, goal)}")
```

## 3. Why use this? (Speed Check)  

The Two-Pointer approach is $O(n)$. This means if you have 10,000 numbers, you only have to look at each number at most once.  

* **Brute Force:** 100,000,000 operations (worst case).  
* **Two-Pointers:** 10,000 operations (worst case).  

The Two-Pointer way is 10,000 times cheaper in operations!

```python
from random import randint, seed
import timeit

seed(55)

big_list = [randint(0, 9999) for i in range(10000)]
big_list.sort()
targ = randint(0, (9999*2)-1)
```

```python
s = timeit.default_timer()
print(find_sum_brute(big_list, targ))

print(f"Brute force took {timeit.default_timer() - s} seconds")
```

```python
s = timeit.default_timer()
print(find_sum_tp(big_list, targ))

print(f"Two-pointer took {timeit.default_timer() - s} seconds")
```

Given an integer array `height` of length `n`. There are `n` vertical lines drawn such that the two endpoints of the `ith` line are `(i, 0)` and `(i, height[i])`.  

Find two lines that together with the x-axis form a container, such that the container has the largest area.  

Return the maximum possible area available.  

Note: the height of the container will be the shorter of the two vertical lines. There will not be a slanted container.

> Explanation: The above vertical lines are represented by array `[1,8,6,2,5,4,8,3,7]`. In this case, the max area of water (blue section) the container can contain is 49.

```python
# def max_area(heights: list[int]) -> int:
#     # hint: if you start the pointers at the ends of the list, you've already maximized the possible width, how can you maximize the height?
#     # ...in other words, when should you move each pointer?  
#     left = 0 
#     right = len(heights) - 1 
#     cur = min(heights[left], heights[right]) * (right - left)
#     while left < right:
#         if heights[left] > heights[right] and left < right: 
#             right -= 1 
#         cur = max(min(heights[left], heights[right]) * (right - left), cur)
#         while heights[left] < heights[right] and left < right:
#             left += 1
#         cur = max(min(heights[left], heights[right]) * (right - left), cur)
#         left += 1 
#         right -= 1
#     return cur 


def max_area(heights: list[int]) -> int:
    # hint: if you start the pointers at the ends of the list, you've already maximized the possible width, how can you maximize the height?
    # ...in other words, when should you move each pointer?  
    left = 0 
    right = len(heights) - 1 
    cur = 0
    while left < right:
        cur = max(min(heights[left], heights[right]) * (right - left), cur)
        if heights[left] > heights[right]: 
            right -= 1 
        else:
            left += 1
    return cur 

```

```python
assert max_area([1,8,6,2,5,4,8,3,7]) == 49, f"Failed, output was {max_area([1,8,6,2,5,4,8,3,7])}"
```
